```python
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, Concatenate
from keras.utils import plot_model
```

```python
main_input = Input(shape=(28, 28, 192))

conv1 = Conv2D(64, kernel_size=(1, 1))(main_input)

conv2_1 = Conv2D(96, kernel_size=(1, 1))(main_input)
conv2 = Conv2D(128, kernel_size=(3, 3), padding='same')(conv2_1)

conv3_1 = Conv2D(16, kernel_size=(1, 1))(main_input)
conv3 = Conv2D(32, kernel_size=(5, 5), padding='same')(conv3_1)

max4 = MaxPooling2D(pool_size=(3, 3), strides=1, padding='same')(main_input)
conv4 = Conv2D(32, kernel_size=(1, 1))(max4)

main_output = Concatenate()([conv1, conv2, conv3, conv4])

model = Model(inputs=main_input, outputs=main_output)
```

```python
model.summary()

plot_model(model, to_file='inception.png', show_shapes=True, show_layer_names=False)
```
