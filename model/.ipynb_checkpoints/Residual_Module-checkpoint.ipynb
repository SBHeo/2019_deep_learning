```python
from keras.models import Model
from keras.layers import Input, Conv2D, Add
from keras.utils import plot_model
```

```python
main_input = Input(shape=(256, 256, 1))
conv1 = Conv2D(64, kernel_size=(1, 1), activation='relu')(main_input)
conv2 = Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu')(conv1)
conv3 = Conv2D(256, kernel_size=(1, 1), activation='relu')(conv2)
main_output = Add()([conv3, main_input])

model = Model(inputs=main_input, outputs=main_output)
```

```python
model.summary()

plot_model(model, to_file='residual.png', show_shapes=True, show_layer_names=False)
```
