```python
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input, Dense, Flatten, Concatenate, Conv2D, MaxPooling2D
```

```python
main_input = Input(shape=(1024,))
dense1 = Dense(128, activation='relu')(main_input)
dense2 = Dense(64, activation='relu')(dense1)
aux_input = Input(shape=(10,))
merge = Concatenate()([aux_input, dense2])
aux_output = Dense(10, activation='softmax')(dense2)
main_output = Dense(1, activation='sigmoid')(merge)
model = Model(inputs=[main_input, aux_input], outputs=[main_output, aux_output])
```

```python
model.summary()
plot_model(model, to_file='model1.png', show_shapes=True, show_layer_names=False)
```
